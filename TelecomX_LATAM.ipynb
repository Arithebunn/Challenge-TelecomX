```python
#Definimos funciÃ³n que quite espacios antes y despuÃ©s del string
def quitar_espacios(df, columnas):
    for col in columnas:

        if df[col].dtypes == 'O':
          df[col] = df[col].astype(str).str.strip().replace('nan',np.nan)

    return df
```

```python
#Definimos funciÃ³n que modifique el contenido de columnas con variables dicotÃ³micas
def yes_no_a_binario(df, columnas):
    for col in columnas:
        valores = df[col].unique()

        if len(valores) == 2 and 'Yes' in valores and 'No' in valores:
            df[col] = df[col].replace({'No': 0, 'Yes': 1}).infer_objects(copy=False).astype(np.int64)

        elif len(valores) == 2 and 'Male' in valores and 'Female' in valores:
            df[col] = df[col].replace({'Female': 0, 'Male': 1}).infer_objects(copy=False).astype(np.int64)

    return df
```

```python
#Definimos funciÃ³n que modifique el contenido de columnas con variables triples
def variable_a_triple(df, columnas):
    for col in columnas:
        valores = df[col].unique()

        if len(valores) == 3 and 'Yes' in valores and 'No' in valores and 'No phone service' in valores:
            df[col] = df[col].replace({'No': 0, 'Yes': 1, 'No phone service': 2}).infer_objects(copy=False).astype(np.int64)

        elif len(valores) == 3 and 'Yes' in valores and 'No' in valores and 'No internet service' in valores:
            df[col] = df[col].replace({'No': 0, 'Yes': 1, 'No internet service': 2}).infer_objects(copy=False).astype(np.int64)

        elif len(valores) == 3 and 'Month-to-month' in valores and 'One year' in valores and 'Two year' in valores:
            df[col] = df[col].replace({'Month-to-month': 0, 'One year': 1, 'Two year': 2}).infer_objects(copy=False).astype(np.int64)

        elif len(valores) == 3 and 'No' in valores and 'DSL' in valores and 'Fiber optic' in valores:
            df[col] = df[col].replace({'No': 0, 'DSL': 1, 'Fiber optic': 2}).infer_objects(copy=False).astype(np.int64)

        elif len(valores) == 4 and 'Mailed check' in valores and 'Electronic check' in valores and 'Credit card (automatic)' in valores and 'Bank transfer (automatic)' in valores:
            df[col] = df[col].replace({'Mailed check': 0, 'Electronic check': 1, 'Credit card (automatic)': 2,'Bank transfer (automatic)': 3}).infer_objects(copy=False).astype(np.int64)

    return df
```

El objetivo de este anÃ¡lisis es **recopilar, limpiar, procesar y explorar los datos de clientes** utilizando Python y sus principales bibliotecas, con el fin de identificar patrones asociados a la cancelaciÃ³n del servicio. ## 2. Limpieza y Tratamiento de Datos

### 2.1 ImportaciÃ³n y normalizaciÃ³n

Los datos se importaron desde un archivo en formato **JSON**, el cual contenÃ­a estructuras anidadas. Para su correcto manejo, fue necesario:
